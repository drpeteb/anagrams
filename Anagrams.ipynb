We'll need a function to load in the word list and parse it, to get rid of apostrophes, make sure everything is the same case, etc.

```python
def load_vocabulary(path):
    """
    Load and parse a word list
    """
    words = set()
    
    with open(path) as fh:
        lines = fh.readlines()
    
    for line in lines:
        word = "".join([letter.lower() for letter in line if letter.isalpha()])
        if word not in exclusions:
            words.add(word)
    
    return words
```

# Efficient Lookups

We can store and reference our word list efficiently by giving each word a key, which is just the sorted letters. When trying to build anagrams, we can save repeated combinations by always requiring that a particular letter (the least frequent is a good choice) be in the next word.

```python
def build_anagram_lookup(words):
    """
    Build a data structure for looking up words.
    """
    lookup = defaultdict(dict)
    
    for word in words:
        length = len(word)
        key = "".join(sorted(word))
        if key not in lookup[length]:
            lookup[length][key] = []
        lookup[length][key].append(word)
    
    return lookup
```

```python
def count_letters(words):
    """
    Build a frequency table of letters.
    """
    frequency = defaultdict(int)
    
    for word in words:
        for letter in word:
            frequency[letter] += 1
    
    return frequency
```

```python
# Build our anagram lookup
vocab = load_vocabulary("gutenberg_clean.csv")
lookup = build_anagram_lookup(vocab)
frequency = count_letters(vocab)
```

# Finding Anagrams

To find anagrams, we will keep track of a list of letters still to be used. In each iteration we will find all the keys which can be formed from these letters. For each of these, we remove the letters used and recurse for each branch.

To avoid stack overflows, we can do this with a queue of partial anagrams. We can avoid repeat effort by storing hashes of the combinations we've checked.

```python
class partial_anagram:
    """
    Represents a partially formed anagram.
    """
    def __init__(self, letters, keys=[]):
        self.keys = set(keys)
        self.letters = sorted([letter for letter in letters], key=lambda letter: frequency[letter])
    
    def copy(self):
        return partial_anagram(self.letters, keys=self.keys)
    
    def hash(self):
        result = 0
        for key in self.keys:
            result = result ^ hash(key)
        return result
    
    def spawn(self, new_key):
        copy = self.copy()
        failed = False
        for letter in new_key:
            if letter in copy.letters:
                copy.letters.remove(letter)
            else:
                failed = True
                break
        if failed:
            copy = None
        else:
            copy.keys.add(new_key)
        
        return copy
```

Once we've worked out all possible combinations of keys which comprise our letters, we need to translate this back into words. We can do this with another recursion.

```python
def make_all_combinations(stem, keys):
    key = keys[0]
    if len(keys) == 1:
        return [stem + " " + word for word in lookup[len(key)][key]]
    else:
        result = []
        for word in lookup[len(key)][key]:
            new_stem = stem + " " + word
            result.extend(make_all_combinations(new_stem, keys[1:]))
        return result
```

```python
def find_anagrams(phrase):
    """
    Find all possible anagrams of the specified phrase,
    using the lookup and frequency data structures provided.
    """
    
    # Valdiate the input
    letters = [char.lower() for char in phrase if char.isalpha()]
    print("Searching for anagrams of: {}".format(letters))
    
    # Set up a hashmap queue to process
    queue = dict()
    first_item = partial_anagram(letters)
    queue[first_item.hash()] = first_item
    
    # Set up a list of results
    results = []
    
    # Set up a record of all the items we've processed
    seen = set()
    
    # Loop through the queue
    while queue:
        
        if len(seen)%100 == 0:
            print("Procesed {} partial anagrams. {} left on queue.".format(len(seen), len(queue)))
        
        # Get a partial anagram from the queue
        item_hash,item = queue.popitem()
        seen.add(item_hash)
        
        # Get the most frequent letter and the number remaining
        letter_must_contain = item.letters[0]
        letters_remaining = len(item.letters)
        
        # Look for keys which can be made from the remaining letters
        for length in range(1,letters_remaining+1):
            for key in lookup[length]:
                if letter_must_contain in key:
                    new_item = item.spawn(key)
                    if new_item is not None:
                        if not new_item.letters:
                            results.append(new_item)
                        else:
                            item_hash = new_item.hash()
                            if (item_hash not in seen) and (item_hash not in queue):
                                queue[item_hash] = new_item
    
    # Now look up all the possible words for the keys
    anagrams = set()
    for item in results:
        for anag in make_all_combinations("", list(item.keys)):
            anagrams.add(anag)
    
    return sorted(list(anagrams), key=lambda anagram: len(anagram.split(" ")))
```

# Trying It Out

```python
tic = time.time()
anagrams = find_anagrams("claude shannon")
toc = time.time()
print("Took {}s to generate anagrams.".format(toc-tic))
```
